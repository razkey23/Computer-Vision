## Output of Cars only

```python
from argparse import ArgumentParser

from mmdet.apis import init_detector, show_result_pyplot
from mmdet.apis import inference_detector_huge_image
```

```python
config = 'configs/obb/oriented_rcnn/faster_rcnn_orpn_r101_fpn_1x_ms_rr_dota10.py'
checkpoint = '/content/drive/MyDrive/Computer-Vision/DOTA/faster_rcnn_orpn_r101_fpn_1x_mssplit_rr_dota10_epoch12.pth'
split = 'BboxToolkit/tools/split_configs/dota1_0/ss_test.json'
img = '/content/OBBDetection/demo/MyImages/lb_3.png'
```

```python
model = init_detector(config,checkpoint,device='cuda:0')
```

```python
# All output classes of model
print(model.CLASSES)
```

```python
nms_cfg = dict(type='BT_nms', iou_thr=0.5) #Used for huge image inference
```

```python
# inference_detector_huge_image() uses a slicing technique for big images (cropout 1024x1024 watch split config)
result = inference_detector_huge_image(model,img,split,nms_cfg)
```

## Remove uninteresting classes

```python
import numpy as np
#Keep only classes that belong to coi (Flexing one-liner)
shapeOfInterest = result[0].shape
coi = {0,4,5,9}
newRes=[res if i in coi else np.empty(shapeOfInterest, dtype=np.float32)  for i,res in enumerate(result)]
```

```python
!python /content/OBBDetection/tools/print_config.py $config
```

```python
model.show_result(img,newRes,score_thr=0.5,out_file='output.jpg') # Result saved in output.jpg
```
