## 2. Inference with a trained YOLOv5 Model

We want to find the ships (if they exist) on the picture above.

We do not use a slicing technique, because the cutout is small and of appropriate size.



- Instantiate a detection model by defining model weight path and other parameters:

````python
detection_model = Yolov5DetectionModel(
    model_path=yolov5_model_ssdd,   #custom model
    confidence_threshold=0.6, #confidence threshold
    device="cuda:0", # or 'cuda:0'
)

````

- Perform prediction by feeding the get_prediction function with an image path and a DetectionModel instance:

````python
result = get_prediction("000111.jpg", detection_model)
````

- Visualize predicted bounding boxes and masks over the original image:

````python
result.export_visuals(export_dir="demo_data/")

Image("demo_data/prediction_visual.png")
````

The predictions are quite good. Our model achieves 98% mAP%0.5 , 95% precision and 96% recall
